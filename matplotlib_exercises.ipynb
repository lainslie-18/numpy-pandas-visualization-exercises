1. Use matplotlib to plot the following equation: 

$y=x^2âˆ’x+2$

You'll need to write the code that generates the x and y points.

Add an anotation for the point `0, 0`, the origin.

```python
x = range(0,18)
y = [(x**2 - x + 2) for x in x]

plt.figure(figsize = (10,8))

plt.plot(x,y,
        c = 'cyan',
        ls = '--',
        marker = '*',
         markerfacecolor = 'b',
        label = 'xy')

plt.ylim(0,300)
plt.xlim(0,17.5)

plt.title('Relationship(x,y)',
         c = 'r',
         fontsize = 15)

plt.grid(True, ls = '--')
plt.annotate('Origin', (0,0), (4,100), arrowprops = dict(), c = 'b', fontsize = 14)
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense)

You can use functions from the `math` module to help implement some of the equations above.

# $y=\sqrt{x}$

```python
x = range(0,50,3)
y = [math.sqrt(x) for x in x]

plt.figure(figsize = (10,8))

plt.plot(x,y,
        c = 'magenta',
        marker = 'o',
        label = 'xy')

plt.title('$y=\sqrt{x}$',
         c = 'b',
         fontsize = 18)

plt.legend()
plt.grid(True)
```

# $y=x^3$

```python
x = range(0,50,3)
y = [math.pow(x,3) for x in x]

plt.figure(figsize = (10,8))

plt.plot(x,y,
        c = '#7E1F0B',
        marker = 'x',
        label = 'xy')

plt.title('$y=x^3$',
         c = '#7E1F0B',
         fontsize = 20)
# plt.yticks([0,20000,40000,60000,80000,100000],[0,20,40,60,80,100])
# plt.ylabel('Thousands')

plt.legend(fontsize = 16)
plt.grid(True, ls = ':')
```

# $y=2^x$

```python
x = range(0,50,3)
y = [math.pow(2,x) for x in x]

plt.figure(figsize = (10,8))

plt.plot(x,y,
        c = '#0E883C',
        ls = 'dashdot',
        marker = 'D',
        label = 'xy')

plt.title('$y=2^x$',
         c = '#0E883C',
         fontsize = 20)

plt.xlabel('x', fontsize = 13)
plt.ylabel('y', fontsize = 13, rotation = 360)

plt.legend(fontsize = 13)
plt.grid(True, ls = ':')
```

# $y=1/(x+1)$

```python
x = range(0,50,3)
y = [(1/(x+1)) for x in x]

plt.figure(figsize = (10,8))

plt.plot(x,y,
        c = '#0C2DD3',
        ls = '--',
        marker = 'd',
        label = 'xy')

plt.title('$y=1/(x+1)$',
         c = '#0C2DD3',
         fontsize = 20)

plt.legend(fontsize = 13)
plt.grid(True, ls = ':')
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = range(0,50,3)

y1 = [math.sqrt(x) for x in x]
y2 = [math.pow(x,3) for x in x]
y3 = [math.pow(2,x) for x in x]
y4 = [(1/(x+1)) for x in x]
```

```python
plt.figure(figsize = (8,6))
plt.suptitle('Subplot Exercise')

plt.subplot(2,2,1)
plt.plot(x,y1,
         c = 'magenta',
         marker = 'o',
         label = 'xy')

plt.title('$y=\sqrt{x}$')
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x, y2,
         c = '#7E1F0B',
         marker = 'x',
         label = 'xy')

plt.title('$y=x^3$')
plt.grid(True, ls = ':')


plt.subplot(2,2,3)
plt.plot(x, y3, 
         c = '#0E883C',
         ls = 'dashdot',
         marker = 'D',
         label = 'xy')

plt.title('$y=2^x$')
plt.grid(True, ls = ':')

plt.subplot(2,2,4)
plt.plot(x, y4,
         c = '#0C2DD3',
         ls = '--',
         marker = 'd',
         label = 'xy')

plt.title('$y=1/(x+1)$')
plt.grid(True, ls = ':')

plt.tight_layout()
```
