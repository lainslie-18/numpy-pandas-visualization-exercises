```python
n = range(1,6)
y1 = [0*n+1 for n in n]
y2 = [math.log(n) for n in n]
y3 = [n for n in n]
y4 = [n*math.log(n) for n in n]
y5 = [n**2 for n in n]
y6 = [2**n for n in n]
y7 = [math.factorial(n) for n in n]
y8 = [n**n for n in n]
```

```python
plt.figure(figsize = (12,10))
plt.title('Big O Notation', c = '#2D8368', fontsize=18)
plt.style.use('seaborn-deep')

plt.plot(n,y1, label='$O(1)$')
plt.plot(n,y2, label = '$O(log\:n)$')
plt.plot(n,y3, label = '$O(n)$')
plt.plot(n,y4, label = '$O(n\:log\:n)$')
plt.plot(n,y5, label = '$O(n^2)$')
plt.plot(n,y6, label = '$O(2^n)$')
plt.plot(n,y7, label = '$O(n!)$')
plt.plot(n,y8, label = '$O(n^n)$')

plt.yscale('log')
plt.xlabel('Elements', c = '#2D8368', fontsize=15)
plt.ylabel('Operations', c = '#2D8368', fontsize=15)
plt.legend(loc = 'best')

plt.text(4.0, 0.12, 'Data Scientist: Lori Ainslie', fontdict=None)
```
