### Scatter plot:  to explore relationship or correlation between two numeric variables

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ravinder79/data/main/insurance.csv')

# assign BMI as x labels for smokers and non-smokers
bmi_smoker = list(df[df.smoker == 'yes'].bmi)
bmi_nonsmoker = list(df[df.smoker == 'no'].bmi)

#assign insurance charges as y labels for smokers and non-smokers
charges_smoker = list(df[df.smoker == 'yes'].charges)
charges_nonsmoker = list(df[df.smoker == 'no'].charges)

```

```python
# plot the data (scatter chart)

plt.figure(figsize = (8,6))

charges_smokers = [charges/500 for charges in charges_smoker]
charges_nonsmokers = [charges/500 for charges in charges_nonsmoker]

plt.scatter(bmi_smoker,charges_smoker, 
            color = 'r', 
            s = charges_smokers,
            label = 'smokers', alpha = 0.4)
plt.scatter(bmi_nonsmoker, charges_nonsmoker, 
            color = 'g', 
            s = charges_nonsmokers,
            label = 'non-smokers', alpha = 0.7)

plt.title('Insurance Charges (USD) vs BMI', fontsize  = 12)
plt.xlabel('BMI')
plt.ylabel('Insurance Charges (USD)')
plt.legend()
```

Conclusions from data:
- Smokers cost more money to insurance
- For smokers, higher the BMI, higher the charges

### Subplots

- Multiple axes (plots) on a single figure
- Use plt.subplot(num_rows, num_cols, index, **kwargs)



```python
# Within a figure - create two subplots( one row and two columns)

plt.figure(figsize = (10,8))

plt.subplot(1,2,1) #1st is num of rows, 2nd is columns, 3rd is index
plt.text(0.5, 0.5, str((1,2,1))), ha = 'centered'

plt.subplot(1,2,2)
plt.text(0.5, 0.5, str((1,2,2))), ha = 'centered'
```

```python
# plot different mathematical functions vs x in a single figure
x = [x * 0.25 for x in range(-25,26)]


y1 = [math.sin(x) for x in x]
y2 = [math.cos(x) for x in x]
y3 = [math.sin(x) * -1 for x in x]
y4 = [math.cos(x) * -1 for x in x]
pi = math.pi

plt.figure(figsize = (8,6))
plt.suptitle('Various Math Functions')

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('$sin(x)$')
plt.xticks([-2 * pi, -1*pi,0, pi, 2*pi]), [$2\pi$,$\pi$,0,$\pi$,$2\pi$]

plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title('$cos(x)$')

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title('$-cos(x)$')

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title('$-sin(x)$')
```

### Histograms: explore distribution of a sample/population

Explore the distribution of a sample/population

```python
# Explore age of respondents from stackoverflow survey
data = pd.read_csv('https://raw.githubusercontent.com/ravinder79/data/main/age_data.csv', index_col=0)
```

```python
plt.figure(figsize = (10, 6))

x1 = [random.gauss(0, 0.8) for _ in range(1000)]
x2 = [random.gauss(-2, 1) for _ in range(1000)]
x3 = [random.gauss(3, 1.5) for _ in range(1000)]
x4 = [random.gauss(7, 0.6) for _ in range(1000)]

# kwargs = dict(alpha = 0.3, bins = 25, density = True)

```

```python
# make subplots using for loop
plt.figure(figsize = (10,8))
x = [x1, x2, x3,x4]
for i in range(0, len(x)):
    plt.subplot(2,2,i+1)
    plt.hist(x[i], ec = 'k', color = 'w')
```

### Summary

Pyplot - Create a plot
- Line plot: plt.plot()
- Scatter plot: plt.scatter()
- Histogram: plt.hist()
- Bar plot: plt.bar()


Pyplot - Customize the plot
- Add title: plt.title()
- Add x or y label: plt.xlabel() or plt.ylabel()
- Add gridlines: plt.grid()
- Add legend: plt.legend()
- Set x and y axis limits: plt.xlim(0,10) or plt.ylim()
- Customize x or y ticks: plt.xticks(), plt.yticks()
- Annotate the plot: plt.annotate()
- Set figure size: plt.figure(figsize=(width, height))
- Draw a horizontal line across axes: plt.axhline()
- Draw a verticle line across axes: plt.axvline()
- Add text on a plot: plt.text()
- plot spacing: plt.tight_layout()

Pyplot - Create Subplots
- plt.subplot(num_rows, num_cols, index, **kwargs)
